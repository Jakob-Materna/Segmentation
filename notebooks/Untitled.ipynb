```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
wing = cv2.imread('../testdata/set_1/043870_L_O.JPG')
```

```python
gray = cv2.cvtColor(wing, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(gray, cmap='gray')
plt.colorbar(shrink=0.45)
plt.axis('on')
plt.show()
```

```python
# wing_denoised = cv2.GaussianBlur(resized_wing, (5, 5), 0)
wing_denoised = cv2.medianBlur(gray, 5)

fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(15, 5), sharex=True, sharey=True)
ax0.imshow(gray, cmap='gray')
ax1.imshow(wing_denoised, cmap='gray')
```

```python
_, binary_image = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10, 10))
plt.imshow(binary_image, cmap='gray')
plt.show()
```
