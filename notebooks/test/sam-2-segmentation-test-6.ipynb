```python
import os
import cv2
import numpy as np
import scipy
import matplotlib.pyplot as plt
from PIL import Image
```

```python
wing = cv2.imread('/home/wsl/proj/master/segmentation/images/high_res_wing_1.jpg')
```

```python
# Show
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(wing, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
r = wing.copy()
# set green and red channels to 0
r[:, :, 1] = 0
r[:, :, 2] = 0

g = wing.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

b = wing.copy()
# set blue and green channels to 0
b[:, :, 0] = 0
b[:, :, 1] = 0
```

```python
# Show
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(20, 20))
ax1.imshow(r)
ax1.set_title("Red")
ax1.axis('off')
ax2.imshow(g)
ax2.set_title("Green")
ax2.axis('off')
ax3.imshow(b)
ax3.set_title("Blue")
ax3.axis('off')
plt.show()
```

```python
bg = wing.copy()
# set red channel to 0
bg[:, :, 0] = 0

# Show
plt.figure(figsize=(5, 5))
plt.imshow(bg)
plt.axis('off')
plt.show()
```

```python
gray_wing = cv2.cvtColor(wing, cv2.COLOR_BGR2GRAY)
gray_r = cv2.cvtColor(r, cv2.COLOR_BGR2GRAY)
gray_g = cv2.cvtColor(g, cv2.COLOR_BGR2GRAY)
gray_b = cv2.cvtColor(b, cv2.COLOR_BGR2GRAY)
gray_bg = cv2.cvtColor(bg, cv2.COLOR_BGR2GRAY)

# Show
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows=5, ncols=1, figsize=(30, 30))
ax1.imshow(gray_wing, cmap="gray")
ax1.set_title("All channels")
ax1.axis('off')
ax2.imshow(gray_r, cmap="gray") 
ax2.set_title("Red channel")
ax2.axis('off')
ax3.imshow(gray_g, cmap="gray") 
ax3.set_title("Green channel")
ax3.axis('off')
ax4.imshow(gray_b, cmap="gray") 
ax4.set_title("Blue channel")
ax4.axis('off')
ax5.imshow(gray_bg, cmap="gray") 
ax5.set_title("Blue and green channel")
ax5.axis('off')
plt.show()
```
