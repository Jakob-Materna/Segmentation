```python
import os
import cv2
import numpy as np
import scipy
import matplotlib.pyplot as plt
from PIL import Image
```

```python
wing = cv2.imread('/home/wsl/proj/master/segmentation/images/high_res_wing_1.jpg')
```

```python
# Show
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(wing, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
r = wing.copy()
# set green and red channels to 0
r[:, :, 1] = 0
r[:, :, 2] = 0

g = wing.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

b = wing.copy()
# set blue and green channels to 0
b[:, :, 0] = 0
b[:, :, 1] = 0
```

```python
# Show
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(20, 20))
ax1.imshow(r)
ax1.axis('off')
ax2.imshow(g)
ax2.axis('off')
ax3.imshow(b)
ax3.axis('off')
plt.show()
```

```python
bg = wing.copy()
# set red channel to 0
bg[:, :, 0] = 0

# Show
plt.figure(figsize=(5, 5))
plt.imshow(bg)
plt.axis('off')
plt.show()
```

```python
gray_1 = cv2.cvtColor(wing, cv2.COLOR_BGR2GRAY)
gray_2 = cv2.cvtColor(r, cv2.COLOR_BGR2GRAY)

# Show
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 20))
ax1.imshow(gray_1, cmap="gray")
ax1.axis('off')
ax2.imshow(gray_2, cmap="gray") 
ax2.axis('off')
plt.show()
```
