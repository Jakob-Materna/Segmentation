```python
import os
import cv2
from pathlib import Path

# Define paths
input_folder = Path("/mnt/c/Users/Jakob/Desktop/tmp/5-LiveWingCells/Wings/")
good_folder = input_folder / "good"
bad_folder = input_folder / "bad"

# Create output folders
good_folder.mkdir(exist_ok=True)
bad_folder.mkdir(exist_ok=True)

# Get a list of all images
image_files = list(input_folder.glob("*.JPG"))  # Adjust extension as needed

for image_file in image_files:
    img = cv2.imread(str(image_file))
    cv2.imshow("Image", img)
    print(f"Processing: {image_file.name}")
    key = cv2.waitKey(0) & 0xFF

    if key == ord('g'):  # Press 'G' for good
        (good_folder / image_file.name).write_bytes(image_file.read_bytes())
    elif key == ord('b'):  # Press 'B' for bad
        (bad_folder / image_file.name).write_bytes(image_file.read_bytes())
    elif key == 27:  # Press 'Esc' to exit
        break

cv2.destroyAllWindows()
```

```python
def all_jpg_files(jpg_dir):
    # Find all jpg files
    jpg_files = []
    for root, _, files in os.walk(jpg_dir):
        for file in files:
            if file.endswith(".JPG") or file.endswith(".jpg"):
                jpg_files.append(file)
    return jpg_files
                
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
    
original_folder = Path("/mnt/c/Users/Jakob/Desktop/tmp/5-LiveWingCells/Wings/original/")
moved_folder = Path("/mnt/c/Users/Jakob/Desktop/tmp/5-LiveWingCells/Wings/moved/")

print(f"original_folder: {len(all_jpg_files(original_folder))}")
print(f"moved_folder: {len(all_jpg_files(moved_folder))}")
print(f"good_folder: {len(all_jpg_files(good_folder))}")
print(f"bad_folder: {len(all_jpg_files(bad_folder))}")

print(f"\n-----\n")
intersect = intersection(all_jpg_files(good_folder), all_jpg_files(original_folder))
print(f"intersection: {len(intersect)}")
```

```python
missing_g = list(set(all_jpg_files(original_folder)) - set(all_jpg_files(good_folder)))
len(missing_g)
```

```python
missing_b = list(set(all_jpg_files(original_folder)) - set(all_jpg_files(bad_folder)))
len(missing_b)
```

```python
intersect = intersection(missing_g, missing_b)
print(f"intersection: {len(intersect)}\n{intersect}")
```

```python
# Copy images from one folder to anotother
import shutil

source_folder = Path("/mnt/c/Users/Jakob/Desktop/tmp/5-LiveWingCells/Wings/original/")
destination_folder = Path("/mnt/c/Users/Jakob/Desktop/tmp/5-LiveWingCells/Wings/")


for file in intersect:
    source = os.path.join(source_folder, file)
    destination = os.path.join(destination_folder, file)

    shutil.copy(source, destination)
```

```python
import pandas as pd

# Define the path to your CSV file
csv_file = "/mnt/c/Users/Jakob/Desktop/tmp/5-LiveWingCells/Data/WingAreas.csv"
out_file = "/mnt/c/Users/Jakob/Desktop/tmp/5-LiveWingCells/Data/FilteredWingAreas.csv"

# Define the list of filenames to keep
filenames_to_keep = all_jpg_files(good_folder)

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file)

# Filter the DataFrame to keep rows where "Filenames" matches the list
filtered_df = df[df["Filename"].isin(filenames_to_keep)]

# Save the filtered DataFrame to a new CSV file (optional)
filtered_df.to_csv(out_file, index=False)

print("Filtered data saved")
```
